## Задание 1.1. Задача о падении тела

**Описание процесса**.
Тело запущено под углом $\alpha$ к горизонту на высоте $h_{start}$ и через

тяжести $${\bf F} = m (0,-g),$$ где $m$ -- масса тела, $g$ -- ускорение
свободного падения.
- Сопротивлением среды пренебрегаем.

- Начальный момент времени движения тела полагаем равным нулю: $t$ = 0.
- Значение массы тела m не повлияет на функции перемещения тела.

**Задания для выполнения**.
- $\color{red}\text{Определите}$ функцию перемещения тела $(s_{x}(t),s_{y}(t))$ под
действием силы тяжести, где $s_{x}(t)$ и $s_{y}(t)$ — горизонтальная и вертикальная
составляющая положения тела в момент времени $t$, соответственно
- $\color{red}\text{Найдите}$ угол запуска тела $\alpha$
- $\color{red}\text{Постройте}$ график траектории движения тела $(s_{x}(t),s_{y}(t))$ для $t ∈ [0,T]$

### Этап 3. Определение функции перемещения и нахождение угла запуска тела


Полагаем, что начало прямоугольной декартовой системы координат соответствует
уровню земли. Тогда в начальный момент движения тела $t = 0$ имеем, что $$s_x(0) = 0, s_y(0) = h_{start}.
$$
Для построения функции перемещения $(s_{x}(t), s_{y}(t))$ воспользуемся вторым
законом Ньютона $$m{\bf a} = {\bf F},
$$
где $m$ — масса тела, ускорение движения тела ${\bf a} = (s
′′
x(t), s
′′
y (t))
$ и приложенная
сила ${\bf F} = m (0,-g),$. В результате имеем два уравнения, что $$s
′′
_x(t) = 0,  s
′′
_y (t) = −g.$$

На основании уравнений с учетом начальных условий можно записать
аналитический вид для функций 
$s_x(t)$ и $s_y(t)$
$$s_x(t) = s_x(0) + v0_xt = 0 + v0_xt,$$
$$s_y(t) = s_y(0) + v0_yt − \frac{g}{2}t^2 = h_{start} + v0_yt − \frac{g}{2}t^2
,
$$


где ${\bf v0} = (v0x, v0y)$ обозначает неизвестную скорость движения тела в начальный
момент времени $t = 0$.

Как изменится вид функций перемещения, если начальный момент времени будет равен не $t = 0$, а $t = t^*$ ?
          




Второй закон Ньютона ($𝐹 = 𝑚𝑎$) показывает, что ускорение тела определяется силой тяжести ($𝑎 = 𝑔$). 
Разложив ускорение на оси, $a_x = 0$, $a_y = -g$, мы через интегрирование находим скорость и положение:

По оси 𝑥:  $s_x(t) = s_x(0) + v0_xt$  
По оси y:  $s_y(t) = s_y(0) + v0_yt − \frac{g}{2}t^2$ 

Если начальный момент времени $t = t^*$, то получим:

По оси 𝑥:  $s_x(t) = s_x(t^*) + v(t^*)_xt$        
По оси y:  $s_y(t) = s_y(t^*) + v(t^*)_yt − \frac{g}{2}t^2$

Это учитывает временной сдвиг, сохраняя зависимость от начальных условий.

Из условия задачи имеем, что $s_x(T) = h_{end}$ . Из этого соотношения находим
значение для $v0_x$

#### Полярное уравнение
a) $p = \frac{a}{m + n sin \varphi},  a \in \mathbb{R}, m \in \mathbb{N}, n \in \mathbb{Z}$


```python
a = 5 
m = 3 
n = 2 

phi = np.linspace(-np.pi, np.pi, 1000)
rho = a / (m + n * np.sin(phi))  

plt.figure(figsize=(8, 6))
ax = plt.subplot(111, polar=True)
ax.plot(phi, rho, label=r"$p = \frac{a}{m + n \sin(p)}$", color="blue")
ax.set_title("Полярный график функции")
plt.legend()
plt.show()
```

#### Параметрическое уравнение


б) $x = \ln \sqrt{\frac{1-t}{1+t}}, \quad y = \sqrt{1-t^2}$

```python
t = np.linspace(-0.99, 0.99, 1000) 
x = np.log(np.sqrt((1 - t) / (1 + t)))  
y = np.sqrt(1 - t**2)  

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r"$x = \ln \sqrt{\frac{1-t}{1+t}}, \, y = \sqrt{1-t^2}$", color="blue")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")  
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")  
plt.title("График параметрических уравнений")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```
