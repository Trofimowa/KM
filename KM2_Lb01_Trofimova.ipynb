–∏–Ω–¥–µ–∫—Å–∞ $k$, –¥–ª—è –∫–æ—Ç–æ—Ä–æ–≥–æ $s_y[k] s_y[k + 1] < 0$. –≠—Ç–æ –±—É–¥–µ—Ç –æ–∑–Ω–∞—á–∞—Ç—å, —á—Ç–æ $
